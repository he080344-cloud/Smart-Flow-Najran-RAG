```python
import torch
import gradio as gr
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# 1. تهيئة البيانات ومحرك البحث المتجه (تمت زيادة k إلى 6 لضمان قراءة اللوائح)
print("جاري تهيئة قاعدة البيانات...")
loader = TextLoader("najran_rules.txt", encoding="utf-8")
documents = loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_documents(documents)
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vector_store = FAISS.from_documents(docs, embeddings)
retriever = vector_store.as_retriever(search_kwargs={"k": 6})

# 2. تحميل النموذج اللغوي (المولد)
print("جاري تحميل نموذج الذكاء الاصطناعي...")
model_id = "Qwen/Qwen2.5-1.5B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto"
)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=200,
    temperature=0.1,
    top_p=0.9,
    repetition_penalty=1.1
)

# 3. دالة معالجة الأسئلة وتوليد الإجابات (تم تحديث الأمر لمنع الهلوسة)
def ask_smart_flow(student_question):
    # حماية النظام من الإدخالات الفارغة
    if not student_question or len(student_question.strip()) < 5:
        return "الرجاء كتابة سؤال أكاديمي واضح لنتمكن من مساعدتك."

    try:
        retrieved_docs = retriever.invoke(student_question)
        context = "\n".join([doc.page_content for doc in retrieved_docs])

        # أمر صارم جداً يمنع التأليف
        prompt = f"""أنت مساعد أكاديمي لجامعة نجران. أجب بناءً على النص المرفق فقط.
قاعدة هامة جداً: لا تقم بتأليف أي عمليات حسابية أو استنتاجات من عندك. إذا سألك عن "أقل عدد ساعات" أو "أكثر عدد ساعات"، ابحث عن كلمة "العبء الدراسي" في النص وأجب بالرقم الموجود هناك فقط وبشكل مختصر جداً.

النص المرفق:
{context}

السؤال: {student_question}
الإجابة المباشرة:"""

        response = pipe(prompt)[0]['generated_text']

        # استخراج الإجابة وتجاوز أخطاء التنسيق
        if "الإجابة المباشرة:" in response:
            final_answer = response.split("الإجابة المباشرة:")[1].strip()
        else:
            final_answer = response.strip()

        # تنظيف الإجابة من أي أسئلة يولدها النموذج تلقائياً
        stop_words = ["السؤال:", "سؤال:", "Question:", "ملاحظة:"]
        for word in stop_words:
            if word in final_answer:
                final_answer = final_answer.split(word)[0].strip()

        return final_answer

    except Exception as e:
        print(f"System Error: {e}")
        return "عذراً، حدث خطأ أثناء معالجة طلبك. الرجاء المحاولة مرة أخرى."

# 4. بناء وتشغيل واجهة المستخدم
print("جاري تشغيل الواجهة...")

# استخدام CSS لدعم اللغة العربية
rtl_css = """
.gradio-container { direction: rtl; text-align: right; }
"""

with gr.Blocks(css=rtl_css) as demo:
    gr.Markdown("# 🎓 المساعد الأكاديمي - سمارت فلو")
    gr.Markdown("مرحباً بك في المساعد الأكاديمي لجامعة نجران. اسأل عن الخطط الدراسية لتخصص نظم المعلومات أو لوائح الحذف والإضافة.")

    with gr.Row():
        user_input = gr.Textbox(label="سؤالك هنا:", placeholder="مثال: كم اقل عدد ساعات ممكنه في الترم الواحد؟", lines=2)

    with gr.Row():
        submit_btn = gr.Button("أسأل سمارت فلو", variant="primary")

    with gr.Row():
        output = gr.Textbox(label="إجابة النظام:", lines=4)

    submit_btn.click(fn=ask_smart_flow, inputs=user_input, outputs=output)

demo.launch(inline=True)
```
